```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
months = ["07/2019", "08/2019", "09/2019", "10/2019", "11/2019"]
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]
df = pd.DataFrame({
    "Month": months,
    "Searches": searches,
    "Direct": direct,
    "Social Media": social_media
})
bar_width = 0.25
x = np.arange(len(df["Month"]))
plt.figure(figsize=(10,6))
plt.bar(x - bar_width, df["Searches"], width=bar_width, label="Searches", color="dodgerblue")
plt.bar(x, df["Direct"], width=bar_width, label="Direct", color="hotpink")
plt.bar(x + bar_width, df["Social Media"], width=bar_width, label="Social Media", color="orange")
for i in range(len(df["Month"])):
    plt.text(x[i] - bar_width, df["Searches"][i] + 1, str(df["Searches"][i]), ha='center')
    plt.text(x[i], df["Direct"][i] + 1, str(df["Direct"][i]), ha='center')
    plt.text(x[i] + bar_width, df["Social Media"][i] + 1, str(df["Social Media"][i]), ha='center')
plt.xticks(x, df["Month"])
plt.xlabel("months")
plt.ylabel("visitors in thousands")
plt.title("Visitors by web traffic sources")
plt.legend()
plt.tight_layout()
plt.show()
```
